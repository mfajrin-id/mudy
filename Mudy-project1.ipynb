```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data10s.csv")
```

```python
x = np.array(data['Konduktivitas'])
x
```

```python
y = np.array(data['Salinitas'])
y
```

```python
fig = plt.figure(figsize = (7, 5))
plt.plot(x, y, 'o', color = 'b')
plt.text(0.45, 0.75, ' y = 0.1089x - 0.1151', transform=fig.transFigure, color='k', fontname="Times New Roman", fontsize = 13.5)
plt.text(0.45, 0.70, '$\mathregular{R^{2}}$ = 0.9977', transform=fig.transFigure, color='k', fontname="Times New Roman", fontsize = 13.5)
plt.xlabel('Conductivity (mS)', fontsize = 15, fontname="Times New Roman")
plt.ylabel('Salinity (ppt)', fontsize = 15, fontname="Times New Roman")
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'y')
plt.savefig('grafik1.png', dpi=600)
```
